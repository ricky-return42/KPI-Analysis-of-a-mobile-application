```python
import numpy as np
x = binom.rvs(n,p,size=1000)
y = binom.rvs(n,0.3,size=1000)
x_mean = x.mean()
y_mean = y.mean()
mean_diff = abs(x_mean - y_mean)
z = np.concatenate((x,y))
for i in range(1000):
    perm = np.random.permutation(z)
    T = abs(perm[:len(x)].mean() - perm[len(x):].mean())
    print T, mean_diff, T>mean_diff
```
